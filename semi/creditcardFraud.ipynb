```python
# 필요 라이브러리 import
import pandas as pd
from sklearn.feature_selection import VarianceThreshold
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import warnings
warnings.filterwarnings('ignore')
```

```python
# seaborn 한글 폰트 설정
# 폰트 경로
font_path = 'C:/Windows/Fonts/Malgun Gothic.ttf'

font_prop = fm.FontProperties(fname=font_path, size=12)
plt.rc('font', family=font_prop.get_name())
```

```python
# 데이터 읽어오기
card_df = pd.read_csv('./creditcard.csv')
card_df
```

```python
# 데이터 info() 확인
card_df.info()
```

```python
# 데이터 describe() 확인
card_df.describe()
```

```python
# target을 X_features랑 y_labels로 분리
X_features = card_df.iloc[:, :-1]
y_labels = card_df.loc[:, 'Class']
X_features.shape, y_labels.shape
```

```python
# target의 분포 확인
print(f'{y_labels.value_counts() / y_labels.count() * 100}')
```

```python
# Variance Threshold 객체 생성
selector = VarianceThreshold(threshold=0.01)

X_features_VT = selector.fit_transform(X_features)
X_features.shape, X_features_VT.shape
```

```python
# 피어슨 상관관계
correlation_matrix = X_features.corr(method='pearson')

# 상관관계 행렬
print(f'상관관계 행렬:\n {correlation_matrix}')

# 상관관계 시각화
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=False, cmap='coolwarm', square=True)
plt.title('피어슨 상관관계')
plt.show()
```
