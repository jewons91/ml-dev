```python
import pandas as pd
import numpy as np
from statsmodels.tools.tools import add_constant
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
# 예제 데이터 ㅅ ㅐㅇ성
np.random.seed(42)
X1 = np.random.rand(100) # 임의의 수
X2 = 2 * X1 + np.random.rand(100) * 0.1 # np.random.rand(100) * 0.1 : 노이즈
X3 = np.random.rand(100)

# 데이터프레임으로 정리
df = pd.DataFrame({
      'X1': X1
    , 'X2': X2
    , 'X3': X3
})
df.head()
```

```python
# 상수항 추가(VIF 계산을 위한 필요)
X = add_constant(df)
X
```

```python
X.shape
```

```python
# VIF 결과를 저장할 빈 데이터프레임 생성
vif_df = pd.DataFrame()
vif_df['feature'] = X.columns
vif_df.head()
```

```python
# vif 계산을 해서 위 데이터프레임에 추가
vif_df['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif_df
```
